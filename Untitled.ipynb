<div class="alert alert-info">
    At this point we've built train_x, and test_x, train_y and test_y will be creating next, but first we clean the categorical variables by substituting numerical values in their place (dummy variables). They are: <br>
    <ul>
        <li>acquisition channel</li>
        <li>sector</li>
        <li>paid_late</li>
    </ul>
</div>

```python
def custom_enumerator(feature, source):
    ret = {}
    i = 0
    for val in source[feature].unique():
        ret.update({val : i})
        i += 1
    return ret
```

```python
acquisition_channel = custom_enumerator('acquisition_channel', train_x)
sector = custom_enumerator('sector', train_x)
paid_late = custom_enumerator('paid_late', train_x)
```

**NB**: We will use these categorical values from the larger dataset (training data) for both training and test data, to prevent independent enumeration which would misalign the two sets.

**NB**: The warnings are ineffective, they are just a Pandas recommendation not to ovveride value, but afects no logic.

```python
train_x['acquisition_channel'] = train_x['acquisition_channel'].map(acquisition_channel)
train_x['sector']    = train_x['sector'].map(sector)
train_x['paid_late'] = train_x['paid_late'].map(paid_late)
```

```python
train_x['acquisition_channel'].unique()
```

```python
test_x['acquisition_channel'] = test_x['acquisition_channel'].map(acquisition_channel)
test_x['sector']    = test_x['sector'].map(sector)
test_x['paid_late'] = test_x['paid_late'].map(paid_late)
```

We need to convert all NaN's to 1 in the employee_count column, defaulting to 1 employee where NaN

```python
train_x['employee_count'] = train_x['employee_count'].fillna(1)
test_x['employee_count'] = test_x['employee_count'].fillna(1)
```

We then fill all other NaN's to 0, in the test set.

```python
test_x = test_x.fillna(0)
```

<div class="alert alert-info">
   Given that test data has a field named payment_status, which train data does not have, and the fact that Numida's performance on this exercise is indicated or driven by wherether the customber pays or not, choosing such a boolean as a target variable is more fitting. Therefore, for <b>train_y</b> and <b>test_y</b> we will use a variable I'll derive below as payment_status for train data, and <b>payment_status</b>, respectively.
</div>

Deriving **train_y** from training data.
Note that, all unrecovered, or partially recovered loans has negative values in the **field cash_yield_15_dpd**. We use this field to classify/infer payment_status. 

```python
def payment_status_mapper(source):
    ret = []
    for i in range(len(source)):
        rec = source.iloc[i]
        if rec['cash_yield_15_dpd'] <= 0: # Written off
            ret.append(1)
        else:
            ret.append(0)
    return ret
```

```python
train_y = payment_status_mapper(train_data)
```

```python
train_y
```

Substituting dummy variables for payment_status in test_y, 0 for paid up, and 1 for Written off.

```python
test_y = test_data['payment_status'].map({'Paid up': 0, 'Written off': 1})
```

```python
test_y = test_y.fillna(1) # Classifying null values as '1', i.e. Written off or not Paid up
```

<div class="alert alert-success">
    We proceed to train our model using a Decision Tree. The choice for a decision tree comes from the obvoius nonlinear relationship of the features, with huge inter variance, and from the fact that the problem has been reduced to a classification problem, and we have a closed problem, meaning no extrapolation needed. We'll use scikit-learn for the Decision Tree.
</div>

```python
dTree = DTC()
```

```python
dTree = dTree.fit(train_x, train_y) # Training the classifier on the training data
```

```python
tree.plot_tree(dTree) # Visualization of the classifier metrics per tree node
```

```python
test_y_pred = dTree.predict(test_x) # Doing predictions based on the test data
```

```python
test_y_pred # Prediction results for each loan, wherether the loan will be paid or written off.
```

```python
sum(train_y), sum(test_y) # Because Written off = 1, a sum = count(written offs)
```

```python
accuracy_score(test_y, test_y_pred) * 100  # Prediction accurary in %
```
